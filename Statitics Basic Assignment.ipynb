# Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples

### Descriptive Statistics
- **Meaning:** Descriptive statistics are methods used to **summarize, organize, and present data** in a meaningful way.  
- **Purpose:** To describe what the data shows without making predictions or generalizations.  
- **Tools/Techniques:**  
  - Measures of Central Tendency: Mean, Median, Mode  
  - Measures of Dispersion: Range, Variance, Standard Deviation  
  - Graphical methods: Pie charts, Histograms, Bar diagrams  

**Example:**  
Suppose we collected the marks of 50 students in a class.  
- Average marks = 65  
- Highest marks = 92  
- Lowest marks = 32  
- Graphical presentation: Histogram of marks distribution  

Here, we are just **describing** the data of 50 students, not making predictions about other classes or populations.  

---

### Inferential Statistics
- **Meaning:** Inferential statistics are methods used to **make predictions, draw conclusions, or test hypotheses** about a larger population based on a sample of data.  
- **Purpose:** To infer or generalize from a sample to a population.  
- **Tools/Techniques:**  
  - Hypothesis Testing (t-test, chi-square test, ANOVA)  
  - Estimation (confidence intervals, standard error)  
  - Regression and Correlation analysis  

**Example:**  
From the same class of 50 students, suppose we want to estimate the **average marks of all students in the school (say 1,000 students)**.  
- We take a random sample of 50 students.  
- Mean score of sample = 65  
- Using inferential statistics, we estimate that the **average marks of all 1,000 students** will be around **65 ± 3** (with 95% confidence).  

Here, we are not only describing the sample but also **drawing inferences about the whole population**.  

---

### Key Difference
- **Descriptive statistics** → Describe and summarize **the data you have**.  
- **Inferential statistics** → Use the data you have to make **conclusions or predictions about a larger population**.  


# Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling

"""
Sampling in Statistics:
Sampling is the process of selecting a subset of individuals or observations from a larger population 
to estimate characteristics of the whole population.
Since studying an entire population is often impractical or costly, sampling helps save time, money, and effort 
while still providing reliable insights.

Random Sampling:
- Every individual in the population has an equal chance of being selected.
- Example: Drawing 50 student names randomly from a school list of 1000 students.
- Advantage: Reduces bias, simple to implement.
- Limitation: May not represent subgroups of the population accurately.

Stratified Sampling:
- The population is divided into subgroups (strata) based on shared characteristics (e.g., age, gender, income).
- Random samples are then taken from each subgroup in proportion to its size in the population.
- Example: Dividing students by grade level, then randomly selecting a proportionate number from each grade.
- Advantage: Ensures all subgroups are represented, leading to more accurate and reliable results.
- Limitation: Requires knowledge of population characteristics in advance.

Key Difference:
- Random sampling selects directly from the entire population.
- Stratified sampling first divides the population into subgroups, then samples from each subgroup.
"""


# Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.

"""
Mean:
- The mean (or average) is the sum of all data values divided by the number of values.
- Example: For [2, 4, 6, 8], mean = (2+4+6+8)/4 = 5.

Median:
- The median is the middle value when the data is arranged in ascending order.
- If there are an even number of values, the median is the average of the two middle values.
- Example: For [1, 3, 5, 7, 9], median = 5. For [1, 2, 3, 4], median = (2+3)/2 = 2.5.

Mode:
- The mode is the value that appears most frequently in a dataset.
- Example: For [2, 3, 3, 4, 5], mode = 3.

Importance of Mean, Median, and Mode:
- These are measures of central tendency, which summarize a dataset with a single representative value.
- Mean is useful for evenly distributed data without extreme outliers.
- Median is useful when the data has outliers or is skewed.
- Mode is useful for categorical data or identifying the most common occurrence.
"""


# Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?

"""
Skewness:
- Definition: Skewness measures the asymmetry of the probability distribution of a dataset.
- Types of Skewness:
  1. Positive Skew (Right Skew):
     - Tail on the right-hand side is longer.
     - Mean > Median > Mode.
     - Indicates presence of high extreme values.
  2. Negative Skew (Left Skew):
     - Tail on the left-hand side is longer.
     - Mean < Median < Mode.
     - Indicates presence of low extreme values.
  3. Zero Skew:
     - Data is perfectly symmetrical (Mean = Median = Mode).

Examples:
- Income distribution is usually positively skewed (a few very high incomes raise the mean).
- Age at death is often negatively skewed (most people live long, but a few die young).

Kurtosis:
- Definition: Kurtosis measures the "tailedness" (peakedness or flatness) of a distribution compared to the normal distribution.
- Types of Kurtosis:
  1. Mesokurtic (kurtosis ~ 3): Normal distribution.
  2. Leptokurtic (kurtosis > 3): Distribution is more peaked with heavy tails (more outliers).
  3. Platykurtic (kurtosis < 3): Distribution is flatter with light tails (fewer outliers).

Positive Skew:
- A positive skew means the data has a longer right tail.
- Most values are concentrated on the left, but a few higher values stretch the mean to the right.
- Example: Income distribution.
"""


```python
# Question 5: Compute Mean, Median, and Mode of a given list

import statistics as stats

# Given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculating Mean, Median, Mode
mean_val = stats.mean(numbers)
median_val = stats.median(numbers)
mode_val = stats.mode(numbers)

# Display results
print("Numbers:", numbers)
print("Mean   :", mean_val)
print("Median :", median_val)
print("Mode   :", mode_val)
```

```python
# Question 6: Compute covariance and correlation coefficient between two datasets (without numpy)

# Given data
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Step 1: Calculate means
mean_x = sum(list_x) / len(list_x)
mean_y = sum(list_y) / len(list_y)

# Step 2: Calculate covariance
cov_xy = sum((x - mean_x) * (y - mean_y) for x, y in zip(list_x, list_y)) / (len(list_x) - 1)

# Step 3: Calculate standard deviations
std_x = (sum((x - mean_x) ** 2 for x in list_x) / (len(list_x) - 1)) ** 0.5
std_y = (sum((y - mean_y) ** 2 for y in list_y) / (len(list_y) - 1)) ** 0.5

# Step 4: Calculate correlation coefficient
corr_xy = cov_xy / (std_x * std_y)

# Display results
print("List X:", list_x)
print("List Y:", list_y)
print("Covariance:", cov_xy)
print("Correlation Coefficient:", corr_xy)
```

```python
# Question 7: Identify outliers in the given dataset (Pure Python)

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Step 1: Sort data
sorted_data = sorted(data)
n = len(sorted_data)

# Step 2: Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = sorted_data[n//4]          # 4th element (18)
Q3 = sorted_data[(3*n)//4]      # 12th element (24)

# Step 3: Compute IQR
IQR = Q3 - Q1

# Step 4: Determine outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 5: Identify outliers
outliers = [x for x in sorted_data if x < lower_bound or x > upper_bound]

# Display results
print("Sorted Data:", sorted_data)
print("Q1:", Q1, "Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound, "Upper Bound:", upper_bound)
print("Outliers:", outliers if outliers else "No outliers")
```

```python
# Question 8: Compute correlation between advertising spend and daily sales (Pure Python)

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Step 1: Calculate means
mean_x = sum(advertising_spend) / len(advertising_spend)
mean_y = sum(daily_sales) / len(daily_sales)

# Step 2: Calculate covariance
cov_xy = sum((x - mean_x) * (y - mean_y) for x, y in zip(advertising_spend, daily_sales)) / (len(advertising_spend) - 1)

# Step 3: Calculate standard deviations
std_x = (sum((x - mean_x) ** 2 for x in advertising_spend) / (len(advertising_spend) - 1)) ** 0.5
std_y = (sum((y - mean_y) ** 2 for y in daily_sales) / (len(daily_sales) - 1)) ** 0.5

# Step 4: Calculate correlation coefficient
corr_xy = cov_xy / (std_x * std_y)

# Display results
print("Advertising Spend:", advertising_spend)
print("Daily Sales:", daily_sales)
print("Covariance:", cov_xy)
print("Correlation Coefficient:", corr_xy)
```

```python
# Question 9: Customer satisfaction survey analysis (Pure Python)

import statistics as stats

# Example survey data (scale 1–10)
survey_data = [8, 9, 7, 6, 10, 8, 7, 9, 8, 6, 5, 7, 8, 9, 10, 6, 7, 8, 5, 6]

# Summary statistics
mean_value = stats.mean(survey_data)
median_value = stats.median(survey_data)
mode_value = stats.mode(survey_data)
stdev_value = stats.stdev(survey_data)

print("Summary Statistics:")
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_value}")
print(f"Standard Deviation: {stdev_value:.2f}")
print()

# Frequency distribution
freq = {i: survey_data.count(i) for i in range(1, 11)}

print("Customer Satisfaction Histogram (Text Version):")
for score, count in freq.items():
    print(f"{score}: {'*' * count}")
```
